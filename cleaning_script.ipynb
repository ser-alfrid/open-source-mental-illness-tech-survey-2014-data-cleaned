# Fixing Age Column
Turning outlier values to NaN and removing non-serious responses

```python
print(df.Age.unique())
print(df.Age.dtype)
```

```python
age_df = df[(df['Age'] > 100) | (df['Age']<18)].copy()
print(age_df.Age.unique())
print(len(age_df.Age.unique()))
```

```python
age_df
```

```python
age_df['Age'] = np.nan
age_drop_columns = [1127]
```

```python
df_altered = df.copy()

df_altered.loc[age_df.index, age_df.columns] = age_df
df_altered = df_altered.drop(age_drop_columns)
df_altered['Age'] = df_altered['Age'].astype('Int64')

df_altered['Gender'] = df['Gender'].map(gender_map).fillna('unspecified')

df_altered.index.name = 'index'
```

```python
df_altered['Timestamp'] = pd.to_datetime(df_altered['Timestamp'])
```
